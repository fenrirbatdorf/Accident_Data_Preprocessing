```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("../Accident_Data_Preprocessing/downsampled_random.csv")
df2 = pd.read_csv("../Accident_Data_Preprocessing/downsampled_sklearnresample.csv")
df3 = pd.read_csv("../Accident_Data_Preprocessing/random_sample_6000.csv")
df4 = pd.read_csv("../Accident_Data_Preprocessing/random_sample_7000.csv")
df5 = pd.read_csv("../Accident_Data_Preprocessing/random_sample_10000.csv")
df6 = pd.read_csv("../Accident_Data_Preprocessing/random_sample_15000.csv")
df7 = pd.read_csv("../Accident_Data_Preprocessing/smotenc.csv")

df = pd.concat([df1,df2,df3,df4,df5,df6,df7], axis = 0)
```

```python
df
```

```python
df0 = df[df['label']==0]
df1 = df[df['label']==1]
df0.shape
```

```python
df1.shape
```

```python
from sklearn.utils import resample
df0_downsamp = resample(df0,
                       replace=True,
                       n_samples=len(df1),
                       random_state=49209)

print(df0_downsamp.shape)

df_downsamp = pd.concat([df0_downsamp, df1])
df_downsamp.reset_index(drop=True, inplace=True)
df_downsamp
```

```python
cols = ['age','type_employer','fnlwgt','education_num','marital_status','occupation','relationship','race','sex','hr_worked_per_week']
df0_downsamp.to_csv("../Accident_Data_Preprocessing/census0.csv", index = None)
df1.to_csv("../Accident_Data_Preprocessing/census1.csv", index  =None)
df_downsamp.to_csv("../Accident_Data_Preprocessing/census_full.csv", index = None)
```

```python
df = pd.read_csv("../Accident_Data_Preprocessing/census_full_binned.csv")
df0 = df[df['label']==0]
df1 = df[df['label']==1]

cols = ['age','type_employer','fnlwgt','education_num','marital_status','occupation','relationship','race','sex','hr_worked_per_week']

df0.to_csv("../Accident_Data_Preprocessing/census0_binned.csv", columns = cols, index = None)
df1.to_csv("../Accident_Data_Preprocessing/census1_binned.csv", columns = cols, index = None)
```

```python
df = pd.read_csv("../Accident_Data_Preprocessing/census_full.csv")
```

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:, :-1]
y = df.label
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2, 
                                                    random_state = 49209,
                                                    stratify = y)

# Export training set to one folder...
X_train.to_csv('../Accident_Data_Preprocessing/X_train_census.csv', index = False)
y_train.to_csv('../Accident_Data_Preprocessing/y_train_census.csv', index = False)

# ...And testing set to another
X_test.to_csv('../Accident_Data_Preprocessing/X_test_census.csv', index = False)
y_test.to_csv('../Accident_Data_Preprocessing/y_test_census.csv', index = False)
```
